# ðŸ•’ Backend Latency Tests

This notebook measures response time for key API endpoints.

```python
import time
import requests

BASE = "http://localhost:8000"

endpoints = [
    ("Chat", "/chat"),
    ("Risk Profile", "/risk_profile"),
    ("Portfolio", "/portfolio?session_id=test"),
    ("Simulation", "/simulate?session_id=test"),
]

for name, ep in endpoints:
    start = time.time()
    try:
        if ep == "/chat":
            requests.post(BASE + ep, json={"session_id": "test", "message": "Hello"})
        elif ep == "/risk_profile":
            requests.post(BASE + ep, json={"session_id":"test","age":30,"income_stability":"high","liquidity_needs":"low","investment_knowledge":"medium","answers":{"q1":3,"q2":4,"q3":2,"q4":5}})
        else:
            requests.get(BASE + ep)
    except Exception as e:
        print(name, "FAILED", e)

    latency = (time.time() - start) * 1000
    print(f"{name} latency: {latency:.2f} ms")
```
