# ðŸ“• BLEU & ROUGE Evaluation for LLM Responses

This notebook calculates BLEU and ROUGE-L for portfolio summaries and explanations.

```python
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu

generated = "Equity funds invest in stocks and offer long-term growth."
reference = "Equity mutual funds primarily invest in stocks for long-term capital appreciation."

# ROUGE-L
scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
rouge_score = scorer.score(reference, generated)

# BLEU
bleu = sentence_bleu([reference.split()], generated.split())

rouge_score, bleu
```
